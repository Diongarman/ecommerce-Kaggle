## Feature Engineering

```python
ecomm_2nd = ecomm_initial.copy()


ecomm_2nd['InvoiceDate'][0].month
#ecomm_2nd.head()


ecomm_2nd['Month'] = ecomm_2nd['InvoiceDate'].apply(lambda date:date.month)
ecomm_2nd['DayOfWeek'] = ecomm_2nd['InvoiceDate'].apply(lambda date:date.weekday())
#ecomm_2nd['DayOfWeek'] = ecomm_2nd['InvoiceDate'].dt.day_name()
ecomm_2nd['Total_Cost_Order_Line'] = ecomm_2nd['Quantity']*ecomm_2nd['UnitPrice']
```

```python
ecomm_2nd.head()

#dayHour = df.groupby(by=['DayofWeek','Hour']).count()['Reason'].unstack()
#dayHour.head()
```

```python
def numberToDay(df):
    
    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
    for key in range(len(df.columns)):


        df = df.rename(columns={df.columns.values[key][1]: dmap[df.columns.values[key][1]]})
    return df
```

## Descriptive Stats (mean, stdev, count)

per month per dayofweek

```python
byMonthDOW_mean = pd.DataFrame(ecomm_2nd.groupby(['Month','DayOfWeek']).mean()['Total_Cost_Order_Line']).sort_values(['Month','DayOfWeek'], ascending=[True,True])
byMonthDOW_std = pd.DataFrame(ecomm_2nd.groupby(['Month','DayOfWeek']).std()['Total_Cost_Order_Line']).sort_values(['Month','DayOfWeek'], ascending=[True,True])
byMonthDOW_count =pd.DataFrame(ecomm_2nd.groupby(['Month','DayOfWeek']).count()['Total_Cost_Order_Line']).sort_values(['Month','DayOfWeek'], ascending=[True,True])

totals = pd.concat([byMonthDOW_mean, byMonthDOW_std, byMonthDOW_count], axis=1)
totals.columns = ['mean', 'std.dev', 'count']

totals
```

## Average Basket(i.e. Order) Spends

### Top 5 highest mean of an order (multiple lines i.e. aggregated invoiceID) cost per country

Netherlands place most expensive orders

Note: UK has highest order count which is why raw spent is highest

```python
mean_spend_by_country = pd.DataFrame(ecomm_2nd.groupby(['Country', 'InvoiceNo'])['Total_Cost_Order_Line'].sum())
# mean_spend_by_country = mean_spend_by_country.groupby('Country').mean().sort_values(by='Total_Cost_Order_Line', ascending=False)[0:6]
# mean_spend_by_country.loc[['United Kingdom']] use this to check counts
# mean_spend_by_country


def f(x):
    d = {}
    d['mean'] = x['Total_Cost_Order_Line'].mean()
    d['count'] = x['Total_Cost_Order_Line'].count()
    d['st.dev'] = x['Total_Cost_Order_Line'].std()
    return pd.Series(d, index=['mean', 'count','st.dev' ])

basket_desc_stats = mean_spend_by_country.groupby('Country').apply(f).sort_values(by='mean', ascending=False)

basket_top_5 = basket_desc_stats[basket_desc_stats['count'] >= 20][0:5] #only show gte 20 baskets, else there is not much to summarise

basket_top_5
```
